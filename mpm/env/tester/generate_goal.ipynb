```python
from diffrl.env.env_hubs import comp_env 
env = comp_env()
env.render('plt')
```

```python
import numpy as np
from diffrl.manual.utils import init_shape
init_shape(
    env,
    {
        'width':  (0.175, 0.175, 0.175),
        'color': [170, 170, 30],
        'method': 'box',
        'center': (0.5, 0.09 + 0.045, 0.5),
    }
)
for i in range(100):
    env.step(np.zeros(12))
env.render('plt')
pos = env.get_state()
init_shape(
    env,
    {
        'radius':  0.1,
        'color': [170, 170, 30],
        'method': 'sphere',
        'center': (0.5, 0.1 + 0.0445, 0.5),
    }
)
goal = env.get_state()['particles'][0]
print(len(goal), len(pos['particles'][0]))

import pickle
with open('sphere.pkl', 'wb') as f:
    pickle.dump([pos, goal], f)

env.render('plt')
```

```python
init_shape(
    env,
    {
        'width':  (0.2, 0.1, 0.1),
        'color': [170, 170, 30],
        'method': 'box',
        'center': (0.5, 0.1 + 0.045, 0.5),
    }
)
for i in range(100):
    env.step(np.zeros(12))

env.render('plt')

pos = env.get_state()
print(pos['particles'][0].mean(axis=0))
init_shape(
    env,
    {
        'width':  (0.1, 0.1, 0.2),
        'color': [170, 170, 30],
        'method': 'box',
        'center': (0.5, 0.1, 0.5),
    }
)
goal = env.get_state()['particles'][0]
env.render('plt')


with open('box.pkl', 'wb') as f:
    pickle.dump([pos, goal], f)
```
